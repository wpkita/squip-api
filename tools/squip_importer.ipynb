```python
import pandas as pd
df = pd.read_excel('moods.xlsx')
df
```

```python
df = df.dropna()
df
```

```python
from datetime import datetime

df['Date'] = (pd.to_datetime(df['Date'] + pd.to_timedelta('22:00:00'))).dt.tz_localize('US/Central').dt.tz_convert('UTC')
df.rename(columns={'Date': 'InstantOccurredAt'}, inplace=True)

utc_now = datetime.utcnow()
df['InstantCreatedAt'] = utc_now
df['InstantUpdatedAt'] = utc_now
df
```

```python
df['Happiness'] *= 10
df['Hopefulness'] *= 10
df['Sadness'] *= 10
df['Pain'] *= 10
df['Guilt/Shame'] *= 10
df['Anger'] *= 10
df['Fear/Anxiety'] *= 10
df['Numb/Empty'] *= 10

df['Happiness'] = df['Happiness'].astype(int)
df['Hopefulness'] = df['Hopefulness'].astype(int)
df['Sadness'] = df['Sadness'].astype(int)
df['Pain'] = df['Pain'].astype(int)
df['Guilt/Shame'] = df['Guilt/Shame'].astype(int)
df['Anger'] = df['Anger'].astype(int)
df['Fear/Anxiety'] = df['Fear/Anxiety'].astype(int)
df['Numb/Empty'] = df['Numb/Empty'].astype(int)
df
```

```python
df = pd.melt(df, id_vars=['InstantOccurredAt', 'InstantCreatedAt', 'InstantUpdatedAt'], value_vars=['Happiness', 'Hopefulness', 'Sadness', 'Pain', 'Guilt/Shame', 'Anger', 'Fear/Anxiety', 'Numb/Empty'], var_name='Mood', value_name='Magnitude')
df
```

```python
moodLookup = {
	"Happiness": "4911fa6e-1400-4f43-b065-ffc8bd98e216",
	"Hopefulness": "4e8668c9-a34a-45ee-977f-aa64ec22d75a",
	"Sadness": "294bd4b4-bf39-4414-ba1e-b3ddf7908687",
	"Pain": "1d679c4b-dd39-46bf-aebe-b93ac548dbaa",
	"Guilt/Shame": "b28de6d7-ec06-4174-8cc4-44232b707bae",
	"Fear/Anxiety": "88128be8-51d0-4dd3-bff4-93202253ffc3",
	"Numb/Empty": "97d70336-4ac7-4b97-919a-3c52349a0f30",
	"Irritable/Agitated": "82b6d3c6-0cfc-432f-b698-9fa1939358d7",
	"Anger": "0b6fed7e-9e16-453c-950d-13f2b272d560"
}
df = df.replace({'Mood': moodLookup})
df
```

```python
df = df.rename(columns={'Mood': 'MoodId'})
df
```

```python
df['UserId'] = '00000000-0000-0000-0000-000000000000'
df['IsArchived'] = False
df
```

```python
import uuid

def generate_guid():
    return uuid.uuid4()

df['Id'] = df.apply(lambda x: generate_guid(), axis=1)
df
```

```python
df = df.reindex(columns = ['Id','InstantOccurredAt','Magnitude','MoodId','IsArchived','InstantCreatedAt','InstantUpdatedAt','UserId'])

df.to_csv('mood-entries-export.csv',index=False)
df.head()
```
